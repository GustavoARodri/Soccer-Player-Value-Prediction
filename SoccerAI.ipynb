#Prediction Current Value Soccer

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("final_data.csv")
df
```

```python
df_treated = df.drop(["player","name","team","position"], axis=1)
```

```python
dummied = pd.get_dummies(df['team'])
```

```python
df_treated = pd.concat([df_treated, dummied], axis=1)
```

```python
dummied = pd.get_dummies(df['position'])
```

```python
df_treated = pd.concat([df_treated, dummied], axis=1)
```

Here a treat the df, i put every unique team as a new column, 0 if he is not that team, 1 if he is in this team, the same with positions

```python
df_treated.info()
```

only float and int, perfect for the model

```python
X = df_treated.drop("current_value", axis=1)
y = df_treated["current_value"]
```

#Training the model

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.ensemble import RandomForestRegressor
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
rfr = RandomForestRegressor(n_estimators=100, random_state = 0)
```

```python
rfr.fit(X_train, y_train)
```

```python
rfr.score(X_train, y_train)
```

```python
y_pred = rfr.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print("R2 score:", r2)
print("RMSE:", rmse)
```

How much R2 score is closer to 1.0 the model is better.

RMSE is the error, how much closer to 0 better, in this case an 3 million error is good model for predict these prices.
